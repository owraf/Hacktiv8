```python
# mengimport library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('london_crime_by_lsoa.zip')
df
```

```python
df.shape
```

```python
df.info()
```

```python
kriminalitas = df[df['value'] >= 1] 
kriminalitas
```

```python
kriminalitas8 = kriminalitas[kriminalitas['year'] == 2008] 
kriminalitas9 = kriminalitas[kriminalitas['year'] == 2009] 
kriminalitas10 = kriminalitas[kriminalitas['year'] == 2010]
kriminalitas11 = kriminalitas[kriminalitas['year'] == 2011]
kriminalitas12 = kriminalitas[kriminalitas['year'] == 2012]
kriminalitas13 = kriminalitas[kriminalitas['year'] == 2013] 
kriminalitas14 = kriminalitas[kriminalitas['year'] == 2014]
kriminalitas15 = kriminalitas[kriminalitas['year'] == 2015]
kriminalitas16 = kriminalitas[kriminalitas['year'] == 2016]
```

```python
sum8 = kriminalitas8['value'].sum() 
sum9 = kriminalitas9['value'].sum()
sum10 = kriminalitas10['value'].sum()
sum11 = kriminalitas11['value'].sum()
sum12 = kriminalitas12['value'].sum()
sum13 = kriminalitas13['value'].sum()
sum14 = kriminalitas14['value'].sum()
sum15 = kriminalitas15['value'].sum()
sum16 = kriminalitas16['value'].sum()
```

```python
df.drop('lsoa_code', axis=1, inplace=True)
df.head()
```

**PERTANYAAN 1** : Berapa jumlah kasus kriminal di London pada tahun 2008 sampai 2016?

**LINE PLOT**

```python
df_1 = kriminalitas[kriminalitas['year'] >= 2008]
df_1
```

```python
df_1.drop(['borough','major_category','minor_category'], axis=1, inplace=True)
df_1
```

```python
plt.plot(['2008', '2009', '2010', '2011', '2012','2013', '2014', '2015', '2016'], 
         [sum8, sum9, sum10, sum11, sum12, sum13, sum14, sum15, sum16])

plt.ylabel('Jumlah Kriminalitas')
plt.xlabel('Tahun')
plt.title('Jumlah kriminalitas di London pada tahun 2008 sampai 2016')
```

**AREA PLOT**

```python
area_array = np.array([['2008', '2009', '2010', '2011', '2012','2013', '2014', '2015', '2016'],
                      [sum8, sum9, sum10, sum11, sum12, sum13, sum14, sum15, sum16]])

area_array = area_array.T
area_array
```

```python
df_1 = pd.DataFrame(area_array, columns = ['tahun', 'jumlah kriminalitas'])
df_1.sort_values('tahun')
```

```python
df_1.set_index('tahun', inplace=True)
```

```python
df_1.index = df_1.index.map(str)
df_1['jumlah kriminalitas'] = df_1['jumlah kriminalitas'].map(int)
```

```python
df_1.plot(kind='area',
          stacked = False,
          figsize=(10,5),
          color='grey',)

plt.title('Jumlah kriminalitas di London pada tahun 2013 sampai 2016')
plt.ylabel('Jumlah Kriminalitas')
plt.xlabel('Tahun')
```

Dari line plot dan area plot diatas dapat disimpulkan bahwa kasus kejahatan di kota London cenderung naik seiring bertambahnya waktu.

**PERTANYAAN 7** : Berapa banyak kriminalitas dengan kategori Common Assault dari tahun 2011 sampai 2016?

```python
df_7 = kriminalitas[kriminalitas['year'] >= 2011]
df_7
```

```python
area_array = np.array([['2011', '2012','2013', '2014', '2015', '2016'], [23898, 23649, 23125, 27763, 30621, 31306]])
area_array = area_array.T
area_array

df_7 = pd.DataFrame(area_array, columns = ['tahun', 'minor_category'])
df_7
```
